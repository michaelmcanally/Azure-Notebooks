```r
raw_df <- read_csv(file = 'insurance.csv')
head(raw_df)
```

```r
model1 <- lm(charges ~ ., data = raw_df)
summary(model1)
```

```r
autoplot(model1)
```

```r
model2 <- lmer(charges ~ age * bmi * children + (1|region) + (1|sex) + (1|smoker), data = raw_df)
```

```r
summary(model2)
```

```r
model3 <- lmer(charges ~ age + bmi + children + (1|region) + (1|sex) + (1|smoker), data = raw_df)
summary(model3)
```

```r
model4 <- lmer(charges ~ age + bmi + children + (1|region) + (1|smoker), data = raw_df)
summary(model4)
```

```r
model5 <- lmer(charges ~ age + bmi + children + (1|sex) + (1|smoker), data = raw_df)
summary(model5)
```

```r
model6 <- lmer(charges ~ age + bmi + children + (1|smoker), data = raw_df)
summary(model6)
```

```r
plot(resid(model6))
```

```r
qqnorm(resid(model6))
qqline(resid(model6))
```
